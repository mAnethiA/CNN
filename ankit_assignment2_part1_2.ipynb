```python
def model_train(model, X, y, epochs = 120, batch_size = 64, criterion = nn.BCEWithLogitsLoss(), optimizer = None):
    model = model
    if optimizer == None:
        optimizer = optim.SGD(model.parameters(), lr=0.01)
    for epoch in range(epochs):
        running_loss = 0.0
        for i in range(0, len(X_train), batch_size):
            #inputs and lables
            inputs = torch.FloatTensor(X[i:i+batch_size].values)
            labels = torch.FloatTensor(y[i:i+batch_size].values.reshape(-1, 1))

            
            optimizer.zero_grad()

            #using bakward and forward
            outputs = model(inputs)
            
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

           #checing loss
            running_loss += loss.item()
        if epoch%10 == 9: 
            print('Epoch %d loss: %.3f' % (epoch + 1, running_loss / len(X)))
```

```python
class NeuralNetwork(nn.Module):
        
        def __init__(self,dropout,activiation, initializer):
            super(NeuralNetwork, self).__init__()
            self.fc1 = nn.Linear(input_size, 128)
            self.fc2 = nn.Linear(128, 64)
            self.fc3 = nn.Linear(64, output_size)
            self.dropout=nn.Dropout(dropout)
            self.activation=activiation
            initializer(self.fc3.weight)
        

        def forward(self, x):
            x = self.fc1(x)
            x = self.activation(x)
            x = self.dropout(x)
            x = self.fc2(x)
            x = self.activation(x)
            x = self.dropout(x)
            x = self.fc3(x)
        
            return x

LRSCH_accuracy = []
model_LRSCH=NeuralNetwork(dropout = 0.1,  activiation = nn.Sigmoid(), initializer = nn.init.uniform_)
criterion=nn.CrossEntropyLoss()
batch_size=64
#max-iterations
epochs=100
#optimizer with learning rate 0.1
learning_rate=0.01

optimizer=optim.RMSprop(model_LRSCH.parameters(), lr=learning_rate)

scheduler=optim.lr_scheduler.StepLR(optimizer, step_size = 9, gamma=0.1)
for epoch in range(epochs):
    for i in range(0, len(X_train), batch_size):
        inputs=torch.FloatTensor(X_train[i:i+batch_size].values)
        labels=torch.FloatTensor(y_train[i:i+batch_size].values.reshape(-1, 1))
        optimizer.zero_grad()
        outputs=model_LRSCH(inputs)
        
        loss=criterion(outputs, labels)
        loss.backward()
        
        optimizer.step()
    scheduler.step()
    if epoch%10==9:
        print('Epoch',epoch+1,'/',epochs,'LR',optimizer.param_groups[0]['lr'])      

    with torch.no_grad():
        inputs_values=torch.FloatTensor(X_val.values)
        labels_values=torch.FloatTensor(y_val.values.reshape(-1, 1))
        
        outputs_values=model_LRSCH(inputs_values)
        
        predictions_values=(outputs_values > 0).float()
        
        accuracy_values=(predictions_values == labels_values).float().mean()
        
        #appending all the values into the list
        LRSCH_accuracy.append(accuracy_values)
print('accuracy testing data:', accuracy_values)
```

```python
def c_gradient(parameters):
    parameters = list(filter(lambda p: p.grad is not None, parameters))
    #defning total_norm
    total_norm = torch.norm(torch.stack([torch.norm(p.grad.detach()) for p in parameters]))
    
    clip_coef = clip_value / (total_norm + 1e-6)
    if clip_coef < 1:
        for p in parameters:
            p.grad.detach().mul_(clip_coef)
        return total_norm
    
clip_value = 1        
GDC_accuracy = []
model_GDC=NeuralNetwork(dropout = 0.2,  activiation = nn.Sigmoid(), initializer = nn.init.uniform_)
criterion=nn.CrossEntropyLoss()
batch_size=64
#max-iterations
epochs=100
#optimizer with learning rate 0.1
learning_rate=0.1

optimizer=optim.RMSprop(model_GDC.parameters(), lr=learning_rate)

for epoch in range(epochs):
    for i in range(0, len(X_train), batch_size):
        #inputs and labels
        inputs=torch.FloatTensor(X_train[i:i+batch_size].values)
        
        labels=torch.FloatTensor(y_train[i:i+batch_size].values.reshape(-1, 1))
        
        optimizer.zero_grad()
        #outputs
        outputs = model_GDC(inputs)
        
        loss = criterion(outputs, labels)
        
        loss.backward()
        
        c_gradient(model_GDC.parameters())
        
        optimizer.step()       

    with torch.no_grad():
        #imputs and lables
        inputs_values=torch.FloatTensor(X_val.values)
        
        labels_values=torch.FloatTensor(y_val.values.reshape(-1, 1))
        
        outputs_values=model_GDC(inputs_values)
        
        predictions_values=(outputs_values>0).float()
        
        gdc_accuracy_values=(predictions_values==labels_values).float().mean()
        #appending all the values into the list
        GDC_accuracy.append(gdc_accuracy_values)
print('accuracy testing data:', gdc_accuracy_values)
```
